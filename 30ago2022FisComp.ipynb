````python



def mi_coseno(x,n_max):
  y=0
  x = x%(2*pi)
  for n in range(n_max):
     y = y + ((-1)**n) * (x**(2*n))/factorial(2*n)
  return y

````

````python
def mi_derivada_h2_old(x,f_x):
  df_x = zeros(len(f_x)-1) # ojo la longitud de df_x es menor que f_x, por eso el -1
  h = x[1] - x[0]
  for i in range(len(df_x)):
    df_x[i] = (f_x[i+1] - f_x[i]) /h
  x_d = linspace(x[0],[-2],len(df_x))
  print(len(x_d),len(df_x))
  return x_d, df_x
````

````python
def mi_derivada_h2(x,f_x):
  df_x = zeros(len(f_x)) # ojo la longitud de df_x es menor que f_x, por eso el -1
  h = x[1] - x[0]
  for i in range(len(df_x)-1):
    df_x[i] = (f_x[i+1] - f_x[i]) /h  # calculo de la derivada hacia adelante
  df_x[-1] = (f_x[-1] - f_x[-2]) /h
  return df_x

def mi_derivada_h3(x,f_x):
  df_x = zeros(len(f_x)) 
  h2 = 2*(x[1] - x[0])
  for i in range(len(df_x)-2):   # puntos interiores
    df_x[i+1] = (f_x[i+2] - f_x[i]) /(h2)
    df_x[0] = (f_x[1]-f_x[0])/(0.5*h2)
  df_x[-1] = (f_x[-1] - f_x[-2]) /(0.5*h2)
  return df_x
````

````python
x = linspace(0,pi,10000)
y = sin(x)
dy = mi_derivada_h2(x,y)
dy_h3 = mi_derivada_h3(x,y)

figure(dpi=100)
plot(x,dy_h3, label = 'f prima numericah3')
plot(x,dy, label = 'f prima numerica')
plot(x,cos(x), label = 'f prima analitica')
legend()
show()
````

````python
# error 
figure(dpi=100)
plot(x,abs(cos(x)-dy))
plot(x,abs(cos(x)-dy_h3))
````

````python
def mi_derivada_h5b(x,f_x):
  df_x = zeros(len(f_x))
  h12 , h = 12*(x[1] -x[0]) , (x[1] - x[0])
  for i in range(len(df_x)-4):
    df_x[i+2]= (-f_x[i+4] + 8*f_x[i+3] - 8*f_x[i+1] + f_x[i])
  df_x = df_x /h12
  df_x[0] = (f_x[1]-f_x[0])/h
  df_x[1] = (f_x[2]-f_x[0])/(2*h)
  df_x[-1] = (f_x[-1]-f_x[-2])/h
  df_x[-2] = (f_x[-1]-f_x[-3])/(2*h)
  return df_x
````
